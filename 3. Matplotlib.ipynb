```python
# Membuat kanvas kosong
fig = plt.figure()

# Menambah axes
# left, bottom, width, and height
axes1 = fig.add_axes([0, 0.6, 0.9, 0.4 ])
axes2 = fig.add_axes([0, 0, 0.9, 0.4 ])

# set titel
axes1.set_title('AXES ONE')
axes2.set_title('AXIS TWO')

# Set plot
axes1.plot(a, b, 'r')
axes2.plot(b, a)

plt.show()
```

```python
# Membuat canvas
fig = plt.figure()

# Mebuat dua axes
# left, bottom, width, height
axes1 = fig.add_axes([0,0, 0.8, 0.8])
axes2 = fig.add_axes([0.1, 0.4, 0.4, 0.3])

# axes1
axes1.plot(a, b, 'b')
axes1.set_title('LARGER PLOT')

# axes2
axes2.plot(b, a, 'r')
axes2.set_title('SMALLER PLOT')

plt.show()
```

```python
# Membuat axes 1 baris 2 kolom
fig, axes = plt.subplots(nrows=1, ncols=2)

# axes pertama
axes[0].plot(a, b, 'r')
axes[0].set_title('First Plot')

# axes kedua
axes[1].plot(b,a, 'g')
axes[1].set_title('Second Plot')

# munculkan plot
plt.show()
```

```python
# Membuat axes 1 baris 2 kolom
fig, axes = plt.subplots(nrows=2, ncols=2)

# axes pertama
axes[0, 0].plot(a, b, 'r')
axes[0, 0].set_title('First Plot')

# axes kedua
axes[0, 1].plot(b,a, 'g')
axes[0, 1].set_title('Second Plot')

# axes ketiga
axes[1, 0].plot(a, b, 'b')
axes[1, 0].set_title('Third Plot')

# axes kedua
axes[1, 1].plot(b,a, 'y')
axes[1, 1].set_title('Fourth Plot')

# Memaksimalkan ruang, sehingga memiliki jarak yang proposional
plt.tight_layout()

# munculkan plot
plt.show()
```

# Figure Size and DPI (dot per inch)

```python
# Mengatur lebar figure (canvas)
# Ukurannya width 3 inches x height 2 inches
# Kepadatan pixel : 200px per inch
# facecolor : Background warna

fig = plt.figure(figsize=(3, 2), dpi=150, facecolor='black')

ax = fig.add_axes([0,0,1,1])

# Label, untuk memberikan keterangans stiap plot pada 'legend'
ax.plot(a , a ** 2, 'r', label='X Squared')
ax.plot(a, a ** 3, 'b', label='X Cubed')

ax.set_title('Title')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')

# Untuk memunculkan 'legend'
# loc : untuk mengatur letak legend, default 'best'
ax.legend(loc='best')

# Menentukan range angka setiap sumbu
ax.set_xlim([0, 5])
ax.set_ylim([0, 150])

# Untuk memunculkan plot
plt.show()
```

```python
# Untuk menyimpan plot sebagai gambar
fig.savefig('./files/my-plot.jpg')
# bbox_inches : agar plot memiliki label, title, dan juga range angka di setiap sumbu
fig.savefig('./files/my-plot-tight.jpg', bbox_inches='tight')
```

# More Customize

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# color : keyword or RGB
# lw (line width) : integer
# ls (line style) : - or -- or -. dsb
# marker : titik setiap nilai : o, +, *
# markersize : integer
# markerfacecolor : keyword or RGB
ax.plot(a, b,
       color='#D0E2EC',
        lw=5,
        ls='-',\
        marker= '*',
        markersize=12,
        markerfacecolor='red'
       )

plt.show()
```
